```python
score = 81

# >=90, grade=A
# >=80, grade=B
#ortherwise grade=F

if score >= 90:
  grade = 'A'
elif score >= 80:
  grade = 'B'
else:
  grade = 'F'
  
  print(f"grade={grade}")
```

```python
# and or
#Java: &&, ||
#Python: and, or
x = 9
if (x > 8 and x < 10):
  print('hello')
  
if(x > 10 or x <= 9):
  print('hello')

#Java:  boolean isElderly = age > 65;
#Python:
age = 67
is_elderly = True if age > 65 else False
# is_elderly = True if age > 65 else 'Peter'  also ok
print(is_elderly)
```

```python
#switch

#after Python 3.10

weekday = 'Wed'
match weekday:
  case 'Mon':
    print('Monday')
  case 'Tue':
    print('Tuesday')
  case 'Wed':
    print('Wednesday')
  case 'Thur':
    print('Thursday')
  case 'Fri':
    print('Friday')
  case _:
    print('default')
    
```
