```python
import statistics
numbers = [100, 90, -5, 70]

def get_stat(nums):
  return max(nums), min(nums), statistics.median(nums), statistics.mean(nums), statistics.stdev(nums)

max_value, min_value, median_value, mean_value, std_value = get_stat(numbers)
print(f"max={max_value}, min={min_value}, median={median_value}, mean={mean_value}, stdev={std_value}")
```

```python
import statistics
numbers = [100, 90, -5, 70]

def get_stats_dict(nums):
  return {"max" : max(nums),"min" : min(nums),"median" : statistics.median(nums),"mean" : statistics.mean(nums),"stdev" : statistics.stdev(nums)}

stats_dict = get_stats_dict(numbers)
print(stats_dict)
```

```python
#Lambda: one time formular

# Java 8 : lambda ->

# Python lambda :
square_formula = lambda x: x**2
print(square_formula(8))

#Java: Stream ->Filter -> collect
numbers = [7,10,9,5,2,100]
even_numbers = list(filter(lambda x: x%2 ==0, numbers)) #stTIC METHOD
print(even_numbers)

names = ['Jennie','Joe','Ben','Lucas']
names_4 = list(filter(lambda x: len(x)>4, names))
print(names_4)
```
