```python
# four types in python

# int x = 3;
x = 3
print(x)
print(type(x))   #type:int

# x = 3.3
x = 3.3
print(x)
print(type(x)) #type:float  #! x's type changed, but still has type

x = 'hello'
print(x)
print(type(x)) #type:str

#pro: free
#con: if you forget the type is changed -> runtime error lol

x = True  #in java: true/false, py:True/False
print(type(x)) #type:bool

#------------------
#String

s = 'hello'
print(s)

s = s+ 'world'
print(s)

s += '!!!'

s2 = "hello"  # double quotation also ok, but '' is more traditional
print(s2)

#String name = "Vincent"
#System.out.println("hello" + name + "!!!")
name = "Vincent"
print(f"hello{name}!!!")  #f declare inside {} is a variable

```

```python
#int float

#problem in Java:
# decimal, divide

result = 10/3    #in java, (int)10/(int)3 = (int)3 
print(result) #3.3333..

result = 10//3  #Floor
print(result)  #3

result = 10//4 #floor
print(result)

#result = 10/0
#print(result) #ZeroDivisionError

# floating point issue
print(0.1+0.2) #0.30000000000000004  #!python still has fpi
print(round(0.1+0.2)) #nearest to integer -> 0
print(round(0.1 + 0.2 , 1)) #nearest to 1 d.p. -> 0.3

radius = 3.5
area = radius * radius * 3.14
print(area)
area = radius ** 2 * 3.14  #/ **->power  in Java:Math.pow()
print(area)

s = 'hello' * 3
print(s)  # hellohellohello  easy :)

y = 3 + 3.0
print(y)
print(type(y))

m = 220000000
print(m) #no overflow

```
