```python
#in python: basic type: int, float, bool, str

names = ['John','Steven','Katie','Lucas', 'Steven']
names.append('Leo')
names.remove('Steven') #only remove the first
print(names)

names.insert(2, 'Peter') #in java, list cannot insert
print(names)

#~Substring in java
print(names[:3])
print(names[1:3])
print(names[4:])

#Java: Collection (List) -> Collections.sort()
names.sort()
print(names)

pop_result = names.pop(1)  #Java, stack<> has pop
print(pop_result)
print(names) #Katie pop out

print(len(names))

names.clear()
print(len(names)) #0

names.append('Andy')
print(names)

print(names[0])
```

```python
#set
#in java new HashSet<>

#python, use bracket symbol to represent data structure
#List -> []
#set  -> {}

ages = {2,9,10,2}
print(ages) #remove dup, orderless, same as java

#add()
ages.add(12)
print(ages)

ages.remove(2)
print(ages)

other_ages = {10,7,5,4}
ages.update(other_ages)
print(ages) #4,5,7,9,10, combine two set, UPDATE ages and remove dup

set1 = {9,4,8}
set2 = {10,4,9,0}
print(set1.intersection(set2)) # retain common items, create NEW set
print(set1)#unchange
print(set2)#unchange

bool_set = {True, False, False, True}
print(bool_set)
```

```python
#!dict

person = {
  "name" : "Steven",
  "age" : 23,
  "height" : 1.76
}
print(person)
print(type(person))

#Java : getter / use key to find value
print(person['age'])
print(person['name'])
print(person['height'])

#Java : HashMap
for k, v in person.items():
  print(f"key={k}, value={v}")
  
  print(person.keys())
  print(person.values())
```

```python
# List of str
# List of dict (List<Object> in java)

members = [
  {
    "name" : "Jennie",
    "age" : 30
  },
  {
    "name" : "Peter",
    "age" : 29
  }
]

print(members)
print(type(members)) #class list

print(members[0]['name']) #Jannie

for e in members:
  print(e['name'])

members[0].update({'name' : 'Steven'})  #alt: members[0]['name'] = 'Steven'
print(members)
```
