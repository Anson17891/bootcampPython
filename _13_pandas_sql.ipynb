```python
import pandas as pd
from sqlalchemy import create_engine, Table, Column, MetaData, Integer, String, Date, Float, BigInteger

db_engine = create_engine("postgresql+psycopg2://postgres:Admin1234@localhost:5432/bootcamp_2510p")

metadata = MetaData()

persons_schema = Table(
  "persons",
  metadata,
  Column("id", BigInteger, primary_key=True), #BigInteger=Long
  Column("name", String(20), nullable=False),
  Column("dob", Date, nullable=False),
  Column("salary", Float, nullable=False)
)

metadata.create_all(db_engine)


df_persons = pd.read_csv("persons.csv")

# write df into DB
df_persons.to_sql("persons", db_engine, index=False, if_exists="replace")

# read from DB
df_sql_result = pd.read_sql("select * from persons where alary > 25000", con=db_engine) #con: connection
print(df_sql_result)
```
