```python
#Java
#Json-> Object Deserialization
#Object -> Json Serialization

import json
#Python 
#Json -> dict
json_str = '{"name" : "Vincent", "age" : 18}'
print(json_str)
print(type(json_str))#class : str

dict1 = json.loads(json_str)
print(dict1)
print(type(dict1)) #class : dict
print(dict1['age']) #18

#dict -> str
json_str = json.dumps(dict1, indent=2) #each attribute add two space
print(type(json_str))
print(json_str)
```

```python
#Read json

with open("user_data.json", "r") as file:
  person_dict = json.load(file) #loads() for str, load() for json

print(person_dict)
print(type(person_dict)) #dict


#Write json file
cat_str = '{"name":"Cathy", "age":10}'
cat_dict = json.loads(cat_str)
with open("cat.json", "w") as file:
  json.dump(cat_dict, file, indent=2)
```

```python
with open("persons.json", "r") as file: #"r" for read
  persons_list = json.load(file)
  
persons_ny = list(filter(lambda x: x['address']['city']=="New York", persons_list))
print(persons_ny)

with open("persons_ny.json", "w") as file:#"w" for write
  json.dump(persons_ny,file, indent=2)
```
