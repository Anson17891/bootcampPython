```python
# list (array)
fruits = ['apple', 'orange', 'lemon']
print(fruits)
print(type(fruits))

#for-each
for fruit in fruits:
  print(fruit)

for fruit in fruits:
  for ch in fruit:  # java : toCharArray()
    print(ch)
    
#for(int i = 0 ; i<5; i++)
for i in range(5):
  print(i)
```

```python
numbers = [10, 8, 3, -4, 7]
#print all odd numbers
for number in numbers:
  if number%2 == 1:
    print(number)
    
word = 'hello world'
#check if 'r' exists
for ch in word:
  if ch == 'r':
    is_r_exist = True
    break

print(f"r exists = {is_r_exist}")

#Alt:
print('r' in word)
```

```python
#while loop 
#Usage in java: LinkedList, Queue/Deque
count = 0
while (count<5):
  count+=1 #no ++
print(count)

# Length
# Java: "hello".length()
print(len('hello'))

names = ['Leo', 'Lucas', 'Oscar']
print(len(names))

#list (get by index)
for i in range(len(names)):
  print(names[i])
```

```python
#Math method
ages = [3, 9, 2, 10, 7]
#max age
#in java: loop
print(max(ages))
print(min(ages))
print(sum(ages))
print(len(ages))

# index of max age
print(ages.index(max(ages)))

#Average age
print(sum(ages)/len(ages))
```
