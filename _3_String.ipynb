```python
word = 'Python'
#Java subString
#in Python, str = char[]
print(word[0])  #P
print(word[-1]) #n   #java: length()-1
print(word[-2]) #o
#print(word[-7]) #IndexError

print(word[:3]) #Pyt #!default starts from 0
print(word[1:3])  #from 1 to 3

print(word[3:])  #default ends with -1

print(word[3:3]) # (nothing)

# method
# Java toUpperCase
print(word.upper()) #PYTHON
print(word.lower()) #python
print(word.replace('y','x')) #Pxthon   not affected by lower() -> return new object

```

```python
# other methods
student = 'Steven'
print(student.count('e')) #2
print(student.isalpha()) #True   #in java: for + check ASCII
print(student.isdigit()) #False  #in java: for + check ASCII
print(student.startswith('S'))#True
print(student.startswith('s'))#False, case sensitive


score = '93'
print(score.isdigit()) #True  #in java: Integer.valueOf() -> throw NumberFormatException
```
