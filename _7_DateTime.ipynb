```python
from datetime import datetime, date, timedelta

print(datetime.now())
print(date.today())

start_date = date(2020,12,31) #create object
print(start_date)

new_date = start_date + timedelta(days=60) # not support month.year
print(new_date)
```

```python
from dateutil.relativedelta import relativedelta

end_date = date.today() + relativedelta(months=3) 
end_date = date.today() + relativedelta(years=3) 
end_date = date.today() + relativedelta(days=3) 
```

```python
formatted_datetime = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print(formatted_datetime)
print(type(formatted_datetime)) #class = str

#change custom format to usual format
timestamp = "2025/10/31 23:59:59"
parsed_datetime = datetime.strptime(timestamp, "%Y/%m/%d %H:%M:%S")
print(type(parsed_datetime)) #class = datetime.datetime
```
