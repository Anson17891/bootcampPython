```python
#dataframe   for data analysis
# series
import pandas as pd

#Series (~Column in excel)

series1 = pd.Series([2,8,-2,10])
print(series1)
print(type(series1)) #pandas.Series

#Dictionary (~HashMap)
staff = {
  "name" : "Leo",
  "age" : 23
}

#Dataframe (~Table)
df1 = pd.DataFrame(staff, index=[0]) #dict->df,  index: add index column, must add when only one row (one scala)
print(df1)

# Java : List<Staff> -> ~a row of data
staff_team = {"names" : ["Leo","Tom","Peter"],  #Column1
              "ages" : [23,18,37]                #Column2
              } 
print(type(staff_team))

df_staff_team = pd.DataFrame(staff_team)
print(df_staff_team)

print(df_staff_team['names'])
print(df_staff_team['ages'])
print(type(df_staff_team)) #pandas.DataFrame
```

```python
#location (loc) (~SQL)
# filter row
df_adult = df_staff_team.loc[df_staff_team["ages"] > 18]  #in sql: select * from dataframe where age >18
print(df_adult)

df_team_2 = df_adult = df_staff_team.loc[df_staff_team["names"].str.len() > 3]   #! str to change item into String
print(df_team_2)

#in if -> and/or
#in df -> &/|
df_team_3 = df_staff_team.loc[(df_staff_team["names"].str.len() == 3) & (df_staff_team["names"].str.contains("L"))]
print(df_team_3)
```

```python
print(df_staff_team.info()) #Columns' property
print(df_staff_team.describe()) # stat of number-column
print(df_staff_team.head()) #first 5 rows
```

```python
# Database/ external source

#In java/Spring: API should be json

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)
print(type(df))
print(df.head())
print(df.describe())
```

```python
#loc

#Female and Allen
df_Allen = df.loc[(df["Sex"] == "female") & (df["Name"].str.contains("Allen"))]
print(df_Allen)



# validate data healthiness (any problematic data)
print(df['Age'].isnull().any()) #True

# to repair data:  delete row / refill N/A or -1 etc
#!Delete row
df_drop_na_age = df.dropna(subset=["Age"])  #! immutable
print(df_drop_na_age['Age'].isnull().any()) #False

print(len(df)) #891
print(len(df_drop_na_age)) #714

#!Refill   
df_for_refill = df.copy()  #in case not want to change origin data
df_for_refill['Age'] = df_for_refill['Age'].fillna("N/A")   #!mutable!
print(df_for_refill['Age'].isnull().any()) #False
print(len(df_for_refill)) #891


# dataframe's stat method
print(df['Age'].mean())   #df_for_refill['Age'] has str(N/A) so cannot use 
print(df['Age'].median())
print(df['Age'].std())
print(df['Age'].var())
print(df['Age'].min())
print(df['Age'].max())

#Group by (~SQL) -> agg function(sum, min, max, avg, count)
df_result = df.groupby("Survived")['Age'].mean()
print(df_result)
```
